```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding, Dropout, BatchNormalization, Input
import tensorflow as tf

# Load your dataset
# Assuming `data` is already loaded in your environment
print(data.columns)  # To verify the columns

# Focus on 'WhiteMoves' to predict the next move
moves = data['WhiteMoves'].astype(str)  # Convert to string if necessary

# Tokenize the chess moves
tokenizer = Tokenizer()
tokenizer.fit_on_texts(moves)

# Convert moves into sequences of integers
sequences = tokenizer.texts_to_sequences(moves)

# Define the maximum length of sequences
max_len = 40  # Adjust according to your data

# Pad the sequences so they all have the same length
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post')

# Split the dataset into training and testing sets (80% training, 20% testing)
X_train, X_test = train_test_split(padded_sequences, test_size=0.2, random_state=42)

# Define vocab_size based on tokenizer
vocab_size = len(tokenizer.word_index) + 1

# Prepare target (next move prediction by shifting sequences)
y_train = X_train[:, 1:]  # Target sequences (next moves)
X_train = X_train[:, :-1]  # Input sequences (excluding last move)

y_test = X_test[:, 1:]  # For testing
X_test = X_test[:, :-1]

# Reshape the target to match the model's expectations (single label per timestep)
y_train = y_train.reshape(y_train.shape[0], y_train.shape[1], 1)
y_test = y_test.reshape(y_test.shape[0], y_test.shape[1], 1)

# Define the LSTM model using Input layer
model = Sequential()

# Input layer with specified input shape
model.add(Input(shape=(max_len - 1,)))  # Shape: (max_len - 1,)

# Embedding layer to map the move sequences to dense vectors
model.add(Embedding(input_dim=vocab_size, output_dim=128))

# First LSTM layer with BatchNormalization and Dropout
model.add(LSTM(units=256, return_sequences=True))
model.add(BatchNormalization())
model.add(Dropout(0.3))

# Second LSTM layer with BatchNormalization and Dropout
model.add(LSTM(units=256, return_sequences=True))
model.add(BatchNormalization())
model.add(Dropout(0.3))

# Third LSTM layer with BatchNormalization and Dropout
model.add(LSTM(units=256, return_sequences=True))
model.add(BatchNormalization())
model.add(Dropout(0.3))

model.add(Dense(len(tokenizer.word_index) + 1, activation='softmax'))  # +1 for padding


# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()

```

```python
# Fit the model to the training data
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=64)

# Evaluate the model on the test data (optional)
test_loss, test_acc = model.evaluate(X_test, y_test)

# Print test accuracy
print(f'Test accuracy: {test_acc}')
```
