```python
import chess
import chess.engine
import random
import pandas as pd

# Load your dataset (replace with the actual path to your CSV file)
#data = pd.read_csv('chess_games.csv')

# Example: Filter games for a specific player
player_name = 'Garry Kasparov'  # You can change this to the player you're interested in
player_games = data[data['White'] == player_name]

# Initialize the chess board
board = chess.Board()

# Load the Stockfish engine
#engine = chess.engine.SimpleEngine.popen_uci("/path/to/stockfish")

# Function to convert algebraic notation to UCI format
def convert_to_uci(board, move):
    try:
        # Attempt to convert the algebraic move to a legal UCI move
        uci_move = board.push_san(move)
        board.pop()  # Undo the move to maintain the game state
        return uci_move.uci()
    except ValueError:
        # If the conversion fails, return None
        return None

# Function to get a move in the style of a specific player
def get_player_move(board, player_games):
    # Stockfish generates possible moves
    result = engine.analyse(board, chess.engine.Limit(time=0.1))
    best_moves = [move.uci() for move in result['pv'][:10]]  # Get the top 10 best moves in UCI format

    # Simulate move selection based on historical moves
    historical_moves = player_games['WhiteMoves']

    # Check if the player has any historical moves and pick one, otherwise use Stockfish's best move
    if not historical_moves.empty:  # Ensure historical moves exist
        historical_move = random.choice(historical_moves.iloc[0])  # Random historical move

        # Convert historical move to UCI format
        uci_move = convert_to_uci(board, historical_move)

        # If the converted move is valid and legal, use it
        if uci_move and chess.Move.from_uci(uci_move) in board.legal_moves:
            return chess.Move.from_uci(uci_move)
        else:
            print("Illegal historical move, using Stockfish's best move")
            return chess.Move.from_uci(best_moves[0])  # fallback to Stockfish's best move
    else:
        # Fallback to Stockfish's best move if no historical moves
        return chess.Move.from_uci(best_moves[0])

# Example of playing a game
while not board.is_game_over():
    # Player's turn
    move = get_player_move(board, player_games)
    if board.is_legal(move):  # Ensure the move is legal
        board.push(move)
    else:
        print("Illegal move selected, using Stockfish's best move")
        move = list(engine.analyse(board, chess.engine.Limit(time=0.1))['pv'])[0]
        board.push(move)

    # Display the current board
    print(board)

# Close the engine when done
engine.quit()
```

```python
import chess
import chess.engine
import random
import pandas as pd

# Step 1: Load the dataset and initialize Stockfish
data_path = r"C:\Users\hiddensardar\Downloads\ches.csv"  # Path to your dataset
data = pd.read_csv(data_path)

engine_path = r"C:\Users\hiddensardar\Downloads\stockfish-windows-x86-64-avx2\stockfish\stockfish-windows-x86-64-avx2.exe"  # Path to your Stockfish engine
engine = chess.engine.SimpleEngine.popen_uci(engine_path)

# Step 2: Player selection
players = [
    "Magnus Carlsen", "Garry Kasparov", "Hikaru Nakamura", "Judit Polgar", 
    "Bobby Fischer", "Anatoly Karpov", "Viswanathan Anand", 
    "Levon Aronian", "Veselin Topalov", "Fabiano Caruana"
]

print("Available Players: ")
for i, player in enumerate(players, 1):
    print(f"{i}. {player}")

# Input the player's choice
player_choice = int(input("Select a player by number: ")) - 1
selected_player = players[player_choice]

print(f"Selected Player: {selected_player}")

# Step 3: Filter the games of the selected player
player_games = data[data['White'] == selected_player]

# Step 4: Initialize the chess board
board = chess.Board()

# Function to convert algebraic notation to UCI format
def convert_to_uci(board, move):
    try:
        uci_move = board.push_san(move)
        board.pop()  # Undo the move to maintain the game state
        return uci_move.uci()
    except ValueError:
        return None

# Function to get a move in the style of a specific player
def get_player_move(board, player_games):
    # Get Stockfish's top 10 moves
    result = engine.analyse(board, chess.engine.Limit(time=0.1))
    best_moves = [move.uci() for move in result['pv'][:10]]

    # Filter historical moves of the selected player
    historical_moves = player_games['WhiteMoves']
    
    if not historical_moves.empty:  # Check if historical moves exist
        # Randomly select a move from historical data
        all_moves = historical_moves.iloc[0].strip("[]").replace("'", "").split(", ")
        historical_move = random.choice(all_moves)

        # Convert to UCI format
        uci_move = convert_to_uci(board, historical_move)
        
        if uci_move and chess.Move.from_uci(uci_move) in board.legal_moves:
            print(f"{selected_player} makes move: {uci_move}")
            return chess.Move.from_uci(uci_move)
        else:
            print(f"Illegal move from {selected_player}'s historical games. Using Stockfish's move instead.")
            return chess.Move.from_uci(best_moves[0])
    else:
        return chess.Move.from_uci(best_moves[0])

# Function to evaluate the game state after every move
def evaluate_board(board, engine):
    info = engine.analyse(board, chess.engine.Limit(time=0.1))
    score = info['score'].relative.score()
    if score is None:
        return "Draw"
    elif score > 0:
        return f"White is winning by {score} centipawns"
    else:
        return f"Black is winning by {-score} centipawns"

# Step 5: Simulate the chess game with alternating turns
while not board.is_game_over():
    print("\nCurrent Board State:\n", board)
    print("\nEvaluating board...")

    # Get the bot's move based on the selected player's style
    move = get_player_move(board, player_games)
    if board.is_legal(move):
        board.push(move)
    else:
        print("Illegal move detected, using Stockfish's best move.")
        move = list(engine.analyse(board, chess.engine.Limit(time=0.1))['pv'])[0]
        board.push(move)
    
    # Display the board
    print(board)

    # Evaluate and print the current board status
    evaluation = evaluate_board(board, engine)
    print(f"\nGame Evaluation: {evaluation}")

# Display the final result
print("\nFinal Board State:\n", board)
print(f"\nGame Over. Result: {board.result()}")

# Close the engine after the game
engine.quit()
```
